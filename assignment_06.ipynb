```python
import numpy as np
import matplotlib.pyplot as plt

# Ask the user to enter a value for M
M = float(input("Enter a value for M: "))

# Generate x values from -10 to 10
x = np.linspace(-10, 10, 400)

# Compute y values
y1 = M * x**2
y2 = M * np.sin(x)

# Plot both functions
plt.figure(figsize=(8, 5))
plt.plot(x, y1, label=r'$y = M \cdot x^2$', color='blue', linestyle='-')
plt.plot(x, y2, label=r'$y = M \cdot \sin(x)$', color='red', linestyle='--')

# Customize plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('Mathematical Functions Plot')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

Q.2 Create a dataset of five subjects and scores.
 Convert it into a Pandas DataFrame.
 Plot the scores using a Seaborn bar plot with:
o Different colors for each bar.
o Annotations on top of each bar.
o Title, axis labels, and grid.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creating the dataset
data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Art'],
    'Score': [88, 75, 92, 81, 95]
}

# Converting the dataset into a Pandas DataFrame
df = pd.DataFrame(data)

# Plotting the scores using a Seaborn bar plot
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x='Subject', y='Score', data=df, palette='viridis')

# Adding annotations on top of each bar
for p in bar_plot.patches:
    bar_plot.annotate(format(p.get_height(), '.1f'),
                      (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha='center', va='center', xytext=(0, 10),
                      textcoords='offset points')

# Adding title, axis labels, and grid
plt.title('Scores by Subject')
plt.xlabel('Subject')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().
 Create a 2x2 subplot layout displaying:
o Line plot showing cumulative sum.
o Scatter plot with random noise.
 Customize with titles, axis labels, and grids.


```python
import numpy as np
import matplotlib.pyplot as plt

# Set your roll number as seed for NumPy (replace 12345 with your actual roll number)
roll_number = 12345
np.random.seed(roll_number)

# Generate a dataset of 50 values using np.random.randn()
data = np.random.randn(50)

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Line plot showing cumulative sum
axs[0, 0].plot(np.cumsum(data), marker='o', linestyle='-', color='b')
axs[0, 0].set_title('Cumulative Sum')
axs[0, 0].set_xlabel('Index')
axs[0, 0].set_ylabel('Cumulative Sum')
axs[0, 0].grid(True)

# Scatter plot with random noise
axs[0, 1].scatter(np.arange(50), data, color='r')
axs[0, 1].set_title('Scatter Plot with Random Noise')
axs[0, 1].set_xlabel('Index')
axs[0, 1].set_ylabel('Value')
axs[0, 1].grid(True)

# Additional empty plots to complete the 2x2 layout (customize as needed)
axs[1, 0].set_visible(False)  # Making this subplot invisible
axs[1, 1].set_visible(False)  # Making this subplot invisible

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

Q.4 Download Data-set from the below link
https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv
Apply ‘seaborn’ library to do the following.
1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset from the given URL
url = 'https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv?raw=true'
data = pd.read_csv(url)

# Display the first few rows of the dataset
print(data.head())

# Extract Total Profit data
total_profit = data['total_profit']

# Line plot for Total Profit
plt.figure(figsize=(10, 6))
sns.lineplot(x=data['month_number'], y=total_profit, marker='o', color='b')
plt.title('Total Profit of All Months')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()

# Extract product sales data
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

# Multiline plot for product sales data
plt.figure(figsize=(12, 8))
for product in products:
    sns.lineplot(x=data['month_number'], y=data[product], label=product)

plt.title('Product Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.legend(title='Product')
plt.grid(True)
plt.show()

# Bar plot for all features/attributes
data.plot(kind='bar', figsize=(14, 8))
plt.title('Company Sales Data')
plt.xlabel('Month Index')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```
